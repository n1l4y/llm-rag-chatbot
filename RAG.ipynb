```python
from newspaper import Article
from langchain.schema import Document

def fetch_text_from_url(url):
    article = Article(url)
    article.download()
    article.parse()
    return article.text

urls = [
    "https://help.testimonial.to/en/articles/6223129-collecting-widget",
    "https://help.testimonial.to/en/articles/6223124-embed-a-single-video"
]
raw_texts = [fetch_text_from_url(url) for url in urls]

documents = [Document(page_content=text) for text in raw_texts]
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=100)
all_documents = text_splitter.split_documents(documents)

print(f"Total number of document chunks: {len(all_documents)}")
```

```python
from langchain.embeddings import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
from langchain.vectorstores import FAISS

db = FAISS.from_documents(all_documents, embeddings)
db.save_local("faiss_index")
```

```python
import torch
import torch.nn as nn

from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("EleutherAI/gpt-neo-2.7B", token="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")

model = AutoModelForCausalLM.from_pretrained(
    "EleutherAI/gpt-neo-2.7B",
    device_map="auto",
    token="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"
)
```

```python
from transformers import pipeline
pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=50
)
```

```python
pip install -U langchain-huggingface
```

```python
from langchain_huggingface import HuggingFacePipeline

llm = HuggingFacePipeline(pipeline=pipe, model_kwargs={'temperature': 0.7})
```

```python
from langchain.chains import RetrievalQA

retriever = db.as_retriever(search_kwargs={"k": 5})
chain = RetrievalQA.from_chain_type(llm=llm, retriever=retriever)
```

```python
def get_response(query):
    response = chain.invoke(query)
    return response

# Example usage
query = "How to embed a testimonial?"
response = get_response(query)
print(response)
```
